```python
num = 1 
while num < 6:
    print('num is currently', num)
    print('num is less than 6, now adding 1 to num')
    num += 1
else:
    print('Now while condition is not True, num is currently', num)
```

```python
num = 1 
while num < 10:
    print('num is currently', num)
    print('num is less than 10, now adding 1 to num, if num != 5')
    if num == 5:
        break
    num += 1
```

```python
# continue key word 
num = 0 
while num < 10:
    num += 1
    if num == 5:
        continue
    else:
        print('num is currently', num)
        print('num is less than 10, now adding 1 to num')
```
